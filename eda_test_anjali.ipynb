```python
import pandas as pd
import zipfile
import io

with zipfile.ZipFile('archive (2).zip', 'r') as z:
    # Assuming 'data/data.csv' is the correct path inside the zip
    # You can check z.namelist() to see all files in the archive
    with z.open('data/data.csv') as f:
        df = pd.read_csv(f)

df.head()
```

1.load dataframe+ missing values + duplication handle

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Basic info
print(df.info())

# Check missing values
print("\nMissing Values:\n", df.isnull().sum())

# Fill missing values (if numeric → median, if object → mode)
for col in df.columns:
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

print("\nDataset shape after cleaning:", df.shape)
```

popularity distribution(histogram)

```python
plt.figure()
plt.hist(df['popularity'], bins=20)
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.title("Distribution of Popularity")
plt.show()
```

3.top 5 least popular tracks

```python
least_popular = df.sort_values(by='popularity').head(5)
print(least_popular[['artists', 'name', 'popularity']])
```

4.most popular track of top 5 artists

```python
import ast

# Safely evaluate the string representations of lists in 'artists' column
df['parsed_artists'] = df['artists'].apply(ast.literal_eval)

# Explode the list of artists to get one artist per row and count occurrences
all_artists = df['parsed_artists'].explode()
top5_artists = all_artists.value_counts().head(5).index.tolist()

print("Top 5 artists:", top5_artists)

for artist in top5_artists:
    top_track = df[df['parsed_artists'].apply(lambda x: artist in x)] \
                .sort_values(by='popularity', ascending=False) \
                .head(1)
    print("\nArtist:", artist)
    print(top_track[['name', 'popularity']])
```
